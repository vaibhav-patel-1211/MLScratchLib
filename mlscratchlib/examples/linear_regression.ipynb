```python
X =  [
  [1],[2],[3]
]
y = [2,3,6]

import numpy as np

X = np.array(X)
y = np.array(y)

model = LinearRegression()
from sklearn.model_selection import train_test_split

X_train,x_test,y_train,y_test = train_test_split(X,y,test_size=0.2)

model.fit(X_train,y_train)
```

```python
pred = model.predict(x_test)
pred
```
